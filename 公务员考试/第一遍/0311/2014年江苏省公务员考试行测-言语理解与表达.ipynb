## 一、言语理解与表达

1.凡是符合立案条件，并以办理立案手续的所有案件，除涉及国家机密或个人隐私等不宜公开的理由外，均应当在人民法院决定立案的7日内在网上公示，同时告知合议庭组成人员和案件审理期限，便于当事人和社会监督，防止拖延案件审理期限，提高司法效率。

这段文字主要说明的是
- A 怎样提高司法效率
- B 加强立案公开制度的措施
- C 案件审理期限的告知方式
- D 司法工作怎样接受社会监督

第一次解题思路：  
由于根据看的粉笔的行测的单节课（袁东）的课，认为抓住关键词【社会监督】，未用中心思想抓题目。   
正确答案B，第一次选择了D

答案解析：

【主旨概括题】
实际上文只有一句话，句子的主干成分为“案件均应在网上公示，同时告知合议庭组成人员和案件审理期限，便于当事人和社会监督”。  
也就是说，案件需要在网上公示同时告知一些相关人员。  
不管是网上公示还是向合议庭组成人员告知都是一种案件的公开，因此B选项正确。

A选项，只是这种公开制度的目的之一  
C选项，告知方式属于文中的部分细节  
D选项，话题错误，文中主要讲的是案件，不是司法工作，过誉笼统。

2. 在中国，让公众理解科学将是一个长期的奋斗目标，除了科学界和教育界的参与，除了更好发挥现有科普资源的作用，眼下更需要做的一件事情是：职业科学普及工作着的培养和扶持，包括一批科普作家队伍的造就。国际上的经验表明，一批职业科学普及撰稿人的存在是科学普及的重要支撑，如果我们能抓紧培养一批职业的科学写作者，或许能为我们的科学普及注入新的动力。

通过此段文字，作者意在   
- A 阐明让公众理解科学是一项长期而艰巨的任务
- B 介绍和评价国际上新的科学普及方面的经验
- C 呼吁尽快培养职业的科普作家和科普工作者
- D 指出科普工作需要科学教育界和公众的共同努力

第一次解题思路：  
抓重点错误，抓了前面的半句话，而不是转折点之后的话。   
选的是D，正确答案是C

答案解析：   
【意图判断题】   
“除了...更...“表示递进关系，“更”之后是文段重点，即“职业科学普及工作者的培养和扶持”是重点，后文通过“国际上的经验”进一步论证了培养职业科学工作者的重要性，这些都是希望能够造就和培养科普专业人才，因此C选项正确。  

A选项，是文段中递进之前的铺垫内容，非重点   
B选项，论证说明文中的内容  
D选项，文中没有提到。

6. 汽车产业是一个长期性产业，作为产品革命带来的核心产品，它形成了相当长的产业链，不仅产业自身的发展会对经济增长具有一定的扩展或成倍的乘数效应，而且会带动金属、电器设备、橡胶等一系列相关产业的发展，极大地解决了就业问题，提升了社会福利，这反过来又会促进汽车工业的进一步发展。

这段文字主要说明的是
- A 汽车产业与其他产业的关系
- B 汽车产业与社会就业问题的关系
- C 汽车产业与产业革命进程的关系
- D 汽车产业与经济社会发展的关系

第一次解题思路：   
排除了bc，觉得d的影响范围太大，从而选择了A。  
正确答案应该是D

答案解析：  
【主旨概括题】   
整个文段是“分-分”结构。  
前半部分讲的是汽车产业发展对经济增长和相关产业的影响。  
后半部分阐述了解决就业和提高福利，即经济社会的发展，对汽车工业的促进作用。   
所以整个文段讲的是汽车产业与经济社会发展的关系，即D选项。

ABC三个选项都是原文中提到过，但是只概括了文段的部分内容，不全面。所以本题选D。

9. 2010年的一项研究显示，参与研究的老年人中，拥有适度社交联系的人比孤独的人活到研究结束时的几率要高出50%。另一份发表于今年的研究，追踪了2000余名中老年人，研究显示，那些自诉孤独感较强的人在6年的研究期间的死亡的比率，是孤独感较低的人的2倍左右。

下列对文意概括最准确的一项是
- A 老年人孤独致死比例较高
- B 适当的社交活动可延长寿命
- C 孤独感较强的人往往不会长寿
- D 孤独感对人的寿命有消极影响

第一次解题思路：  
排除B，C。由于D在原文中没有直接的表达，是引出的影响，处于保守考虑，排除了D，选择了A。 实际上应该选择D。

答案解析：  
【主旨概括题】   
并列结构。  
文章列举了两个事例，第一个事例证明了孤独的人活的相对短。  
第二个事例证明自诉孤独感较强的人的死亡率比较高。  
两个事例都能证明孤独感对寿命有所影响。

A太绝对，孤独只能影响寿命，但不一定导致死亡。  
B只跟第一个事例有关联。  
C只跟第二个事例有关联，都是片面说法，且过于绝对。

所以选D
