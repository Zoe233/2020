<h3><a name='combine'>1.组合排列</a></h3>

题目特征：
- 题干给出两组或两组以上对象
- 给出几组对象之间的关系

考点一：排除法、代入法
- 排除法
    - 当题干条件为真时，使用
    - 读一句，排一句
- 代入法：
    - 题干条件有真有假
    - 把选项代入题干去验证

例如：   
甲、乙、丙 三个人，一个是美国人，一个是中国人，一个是日本人。中国人比乙年龄小，甲和中国人不同岁，丙的年龄比日本人大。


根据上述断定可以推出：
- A 甲是日本人，乙是美国人，丙是中国人
- B 甲是美国人，乙是中国人，丙是日本人
- C 甲是中国人，乙是日本人，丙是美国人
- D 甲是中国人，乙是美国人，丙是日本人

答案解析：
- 通过中国人比乙年龄小，得出乙不是中国人，排除B
- 通过甲和中国人不同岁，得出 甲不是中国人，排除C和D  
正确答案为A

考点二：辅助技巧
- 最大信息：以**出现次数最多的信息**为推理起点
- 列表格：
    - 当题干信息存在比较的时候，用"<",">"串起来
    - 当题干信息关系复杂的时候，可以列二维表

例如：  
甲、乙、丙、丁四人的职业是：教师、医生、设计师、演员。其中：  
（1） 甲不是教师，也不是演员   
（2） 乙不是设计师，也不是教师  
（3） 如果甲不是设计师，那么丁不是教师  
（4） 丙不是演员，也不是教师。   

根据以上陈述，可知下列哪项为真？  
- A 甲是设计师，乙是医生
- B 丙是教师，丁是演员
- C 甲是设计师，丁是医生
- D 乙是教师，丙是医生

答案解析：选D  
- 根据（1）（2）（4）条件，可以针对条件（3），得到 丁一定是教师。
-

|甲|NO||推理ok|NO|
|乙|NO||NO|推理ok|
|丙|NO|推理ok||NO|

例如：  
短跑决赛中，前三名甲、乙、丙是A、B、C队的选手。  
已知：  
- A队选手的成绩比B队选手的成绩好
- C队选手的成绩比乙差
- C队选手的成绩比丙好

根据上述条件，下列选项中，（）项肯定为真。  
- A 甲、乙、丙依次为C队，B队和A队选手
- B C队选手是冠军，A队选手是亚军，B队选手是季军
- C 甲、乙、丙此次为C队，A队和B队选手
- D A队选手是冠军，B队选手是亚军，C队选手是季军

答案解析：  
- 根据条件的C队选手的成绩得到 甲就是 C队。
    - 得到 甲比乙差，甲比丙好
    - 成绩是：乙，甲 ，丙
- 因为A选手成绩比B好，则 A 在 B前面，则乙是A队，丙是 B队的。
- 正确答案选C

错题：   
某高校旅游管理专业学生组织甲、乙、丙、丁、戊5支队伍，暑期赴安徽、山西、湖北、四川、广西5省调研旅游景点管理现状。他们各自均有一个特定的调研对象，分别是寺庙、道观、园林、海滩及村落，没有遗漏也没有重复。已知：
（1）甲队比调研寺庙的队伍出发早，但比调研村落的丙队出发晚；
（2）如果调研寺庙的不去山西，则调研海滩的不去广西；
（3）调研园林的乙队出发最早；
（4）调研海滩的丁队出发最晚。
根据以上信息，可以得出以下哪项？
- A 甲队赴四川调研道观
- B 乙队赴湖北调研园林
- C 丙队赴安徽调研村落
- D 戊队赴山西调研寺庙

答案解析：  
- 第一步，分析题干
    - （1）（2）（3）（4）可以得出甲去道观，乙去园林，丙去村落，戊去寺庙，丁去海滩；无法排除选项
    - 根据（2）可知：去寺庙的 戊 不去山西，去海滩的 丁 不去广西
    - 5个省份只有 广西 有海滩（此处需要补充常识）！！！（艹，就是这个导致我做不出来）
        - 所以 丁 一定去广西，根据否后推否前，则去寺庙的戊 一定是去山西，得到答案D
    
    

错题：   
有三位男生张强、赵林、王刚和三位女生李华、秦珊、刘玉暑假外出旅游，可能去上海、杭州、青岛和大连。  
条件是：  
①每人只能去一个地方；  
②凡是男生去的地方就必须有女生去；    
③凡是有女生去的地方就必须有男生去；    
④李华去上海或者杭州，赵林去大连。

如果上述断定都为真，则去杭州的人中不可能同时包含（    ）。
- A 张桥和王刚
- B 王刚和刘玉
- C 秦珊和刘玉
- D 张强和秦珊

答案解析：  
根据条件2和3可知，任何一个地点必须是男女搭配，四个条件中唯一能确定的是 赵林是大连，根据“确定信息优先“的原则，先从赵林开始推理。  
- 赵林是男生，去大连，那么三个女生中至少有一个要去大连，而根据4⃣️，李华去上海或杭州，不去大连，所以去大连的女生 要么是 秦珊，要么是刘玉，所以这两个女生不可能同时去杭州。
- 故答案为C。
