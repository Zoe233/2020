（单选题）发动机：飞机：旅客 （）
- A 琴键：钢琴：音乐
- B 电池：手电筒：灯光
- C 发条：钟表：时间
- D 杆秤：秤杆：重量

答案解析：
- 第一步：判断题干词语间的逻辑关系。
    - 发动机 是 飞机 的组成部分，且是 飞机 的动力来源装置，飞机通过运载 乘客 实现其功能
- 第二步：判断选项词语间的逻辑关系。
    - 琴键 是 钢琴 的组成部分，但是 琴键 不是飞机的 动力来源装置，排除
    - 电池 并非是 手电筒的组成部分，而是 配套关系 的对应，排除（第一次做错误地将其认为是动力来源）
    - 发条 是 钟表 的组成部分，也是 钟表 的动力来源装置，钟表通过计算 时间 实现其功能
    - 秤杆 是 杆秤 的组成部分，但词语顺序不一致。排除
    
正确答案为C。

（单选题）桎梏：刑具：束缚（）
- A 月光：月亮：皎洁
- B 曙光：阳光：希望
- C 兄弟：手足：友谊
- D 危机：风险：突发

答案解析：
- 第一步：判断题干词语间的逻辑关系。
    - 桎梏 是中国古代的一种刑具，在手上戴的为梏，在脚上戴的为桎，可以象征束缚。前两者是 种属关系，第一个词和第三个是为 喻象征义。
- 第二步：判断选项词语间的逻辑关系。
    - 月光不是月亮，不是种属关系，排除
    - 曙光是破晓时的阳光，用来比喻希望。曙光属于阳光的一种，象征希望。正确
    - 兄弟比喻成手足，不是种属关系，排除
    - 危机具有风险性和突发性，后两个词与第一个词是属性对应，排除。
    
正确答案为B

<h3><a name='data'>3.资料分析（简单计算，直接找数，排序类）</a></h3>



注意坑：
- 时间
    - 时间点、时间段、同比/环比
- 单位
    - 亿/万亿，%/千分之一，元/美元
- 方向
    - 最多/最少，从高到低/从低到高，高于/低于

（单选题）2015年1-10月份，第三产业中，基础设施投资（不含电力）79510亿元，同比增长17.4%，增速比1-9月份回落0.7个百分点。其中，水利管理业投资增长23%，增速提高0.5个百分点；公共设施管理业投资增长19.4%，增速回落0.6个百分点；道路运输业投资增长17.7%，增速回落0.4个百分点；铁路运输业投资增长1.4%，增速回落0.4个百分点。


2015年1-9月份第三产业的基础设施投资（不含电力）中，下列项目同比增速从大到小排序正确的是（）
- A 水利管理业投资，公共设施管理业投资，道路运输业投资，铁路运输业投资
- B 铁路运输业投资，道路运输业投资，公共设施管理业投资，水利管理业投资
- C 水利管理业投资，道路运输业投资，公共设施管理业投资，铁路运输业投资
- D 铁路运输业投资，公共设施管理业投资，道路运输业投资，水利管理业投资

题目解析：！！！
- 注意：材料提供的是1-10月份的，题目中问的是1-9月份的。
- 10月份的同比增长17.4%，较1-9月份回落0.7，即1-9月份的同比增长为 （17.4+0.7）%。
    - 以此类推：
        - 水利：（23-0.5）=22.5%
        - 公共：（19.4+0.6）=20%
        - 道路：（17.7+0.4）=18.1%
        - 铁路：（1.4+0.4）=1.8%
        
正确答案选A。   
虽然哪怕审题错误也不会影响答案，但是这个是对于我个人的材料题的盲区的解惑。

<h3><a name='order'>4.语句排序</a></h3>

提问方式：  
将以上几个句子重新排列，语序正确的是...

<img src='./images/010.png' width=300></img>

理论要点：
- 对比选项，确定首句
- 确定捆绑集团/确定顺序/确定尾句
- 验证

**确定首句**：
- 适合做首句
    - 背景引入
        - 随着，近年来，在...大背景/大环境下，长期以来
    - 下定义
        - ...是，被称为...
        - ...就是/是指
    - 提出观点
        - 有人说，人们普遍认为
- 不适合做首句
    - 关联词后半部分，指代词，由“如”引导的例子

**确定捆绑**：
- 关联词捆绑
    - 配套出现（不但，而且）
    - 单独一个（但，同时，分析句子意思）
- 指代词捆绑
    - 这，那，他，该，其
- 共同信息捆绑
    - 着特殊信息捆绑

（单选题）①魏晋时期是书家辈出的时代   
②尺牍之迹因书法而重于世的情况越来越普遍   
③许多书家都擅长尺牍之学   
④魏晋时期，“工尺牍”和“善书”被视为值得称道的修养和才能，二者关系密切  
⑤书家写的尺牍尤其具有收藏价值   
⑥这种情况在魏晋以来的史籍里得到反映，比如“言书法则不离尺牍”     
将以上6个句子重新排列，语序正确的是：  （）
- A 1，3，4，2，5，6
- B 2，3，5，1，6，4
- C 3，4，5，6，1，2
- D 4，6，1，3，2，5

答案解析：  
- 观察选项，首句不易判断，四句话都可以作为首句特征。    
- 再观察句子，6⃣️句中，出现“这种情况”，考虑 **指代词捆绑**，6⃣️句中的举例“言书法则不离尺牍”，强调二者关系的密切，4⃣️句也强调二者关系的密切，故4⃣️6⃣️捆绑，选择D项。

说实话，这题我觉得每句话都挺恶心，完全靠4⃣️6⃣️捆绑。即使是D项。也不是完全通顺的感觉。

（单选题）将下面5个句子重新排列，语序正确的是（  ）。   
①商代以后，随着文字的出现，书写需求增加，周宣王时期“刑夷始制墨”，出现了颗粒状人工墨  
②到了以彩陶为典型特征的仰韶文化时期，已经有了用来研墨颜料的研磨器——石研  
③这些长过半米的大石器，是用来加工粮食的  
④砚起源于研磨器，目前已知最早的研磨器是距今七、八千年的裴李岗、磁山文化时期的石磨盘和石磨棒   
⑤此后，用于研磨的石研，就成为早期的砚  

- A 1，3，4，5，2
- B 4，3，2，1，5
- C 1，3，4，2，5
- D 4，3，2，5，1

答案解析：

- 从选项入手，首句不好判定
- 3⃣️句中出现 代词”这“，”这些...大石器”指代4⃣️句中的“石磨盘和石磨棒”，排除AC。
- 对比BD两项，不同之处在于1⃣️5⃣️的顺序，1⃣️句讲的是出现了墨，5⃣️句讲的是出现了砚，按照先后顺序，应该先有墨，才有研墨的砚，所以1⃣️5⃣️相连。

从上面两道题可以看出，如果实在通读不符合语感，则通过方法论来判断。

（单选题）
①作为新兴的农村商业中心，市镇并不以政治中心为依托，而是基于商业和市场活动兴起，又以商品经济体的形式发展，因而具有鲜明的经济中心的特征

<h3><a name='tiankong'>5.逻辑填空</a></h3>




- **转折**
    - 标识词
        - 但是，然而，却...
    - 理论要点
        - 前后语义相反
- **递进**
    - 标识词
        - 而且、甚至、更...
    - 理论要点
        - 语义程度前轻后重

- 关联关系：**并列**
    - 理论要点：
        - 同义并列：顿号（、），逗号（，）
            - 如：面对反腐，我们要有踏石留印、抓铁有痕的决心。
        - 反义并列：不是...而是...,是...不是...,相反，反之...
            - 如：乐观的人...,反之，悲观的人...

例题： 
目前，我国不少大城市已处于超载状态，“城市病”严重，而人口还在不断 ________ ，许多中小城市和小城镇发展缓慢，甚至出现 ________ ，对农名工失去吸引力。面对这一现状和城镇化所带来的巨大压力，城市群的发展被寄予厚望。
- A 增加 空置
- B 迁移 停滞
- C 涌入 萎缩
- D 流动  退化

答案解析： 选C 
- 根据超载状态，排除BD
- 根据发展缓慢，递进关系，选择C

-----这种题目我的准确率并不高的原因，不是因为我逻辑不好，而是词语的意思经常性无法判断正确，还是高中语文的那套词语的准确性训练-----

- 词义侧重
    - 【辨析方法】
        - 用不一样的字组词（注意：组词不要离原词意思差太远）
        - 整词进行 固定搭配
            - 例如：品行，品性
                - 品行：行为，行动，外在
                - 品性：个性，性格，内在
               

- 固定搭配
    - 常用词搭配、热点词搭配
        - 爱护/爱戴
        - 经济新常态 供给侧结构性改革 京津冀协同发展
    - 找准搭配对象（瞻前顾后）

- 对应关系——解释类对应
    - 题干特点：分句，_____ ，分句
    - 标识词：是、就是、即、无异于、无疑是、比如、例如等
    - 标点：冒号（：）、破折号（——）

<h3><a name='jiaqiang'>7.判断：加强论证</a></h3>




常见提问方式：
- 如果以下各项为真，最能 加强/支持/赞同/证明 上述论断的是？
- 以下哪项为真，是上述论断成立的前提/假设/必要条件？
- 如果以下各项为真，最不能加强上述论断的是？

- 加强的方法
    - 搭桥
        - 何时
            - 当 **论点和论据 讨论的话题不一致**
            - 提问方式问 前提、假设、必要条件、支持论证 时，优先考虑搭桥
        - 例如：
            - 楼市成交量在增长，因此楼市价格会上涨
                - 分析：论据是：楼市成交量在增长  论点是：上涨  可以用 “搭桥”
                - 解：楼市成交量的增长会带动楼价的上涨；（优选答案）
                - 解：楼市成交量会影响楼市价格（次选答案）
            
    - 补充论据
        - 必要条件：选项为论点成立的必要条件（没它不行）
            - 特征：
                - 提问方式问前提、假设、必要条件、支持论证
                - 而且没有 搭桥项
        - 解释：说明论点成立的原因
            - 特征：
                - 论点和论据之间话题无明显不一致
                - 题干中没有论据，或者论据只是背景描述
           
        - 举例：证明论点成立的例子

例题：   
有人认为，S国政府对社会科学研究的经费投入不见得有利于科研水平的提高，近10年间，S国政府对社会科学研究的经费投入年增长超过20%。但是这十年间S国社会科学领域发表的论文数量却没有显著增长。   

上述结论的成立必须补充以下哪一项前提？（）
- A 论文发表数量是衡量社会科学研究水平的重要指标
- B 近十年间S国社会科学领域的研究人员数量显著增长
- C 近十年间S国物价水平逐年增长
- D 近十年间S国社会科学领域的研究人员在国外发表的论文数量没有显著增长

答案解析：  选A   
- 通过”加强-搭桥法“
    - **论据**：经费投入年增长超过20%，但发表论文的数量没有显著增长
    - **论点**：经费投入，不见得有利于科研水平的提高
- 首先BC是削弱论点，排除
- **A是说明，论文数量 是 科研水平的指标，可以通过论文数量判断科研水平发展 <== 这种就是“搭桥”加强**
- D是和原论据同样的论据，只是说明了在国外的情况 无法增强   
选A

<h3><a name='jiqixianqi'>8.资料分析（基期现期）</a></h3>




#### 基期量
- 识别：给现在，求以前某时期的值
- 公式：$$基期量=现期量-增长量$$
$$基期量=\frac{现期量}{1+r}$$
- 速算：
    - |r|>5%，截位直除
    - |r|<=5%，化除为乘
        - 【化除为乘】
            - 何时用：求基期，选项差距小，|r|<=5%
            - 怎么用：
                - $\frac{A}{1-r} \simeq A * (1+r)$
                - $\frac{A}{1+r} \simeq A * (1-r)$

```python
print(1000/1.03)
print(1000 * 0.97)
```

#### 现期量
- 识别：给现在，求以后某时期的值
- 公式：$$现期量=基期量+增长量$$
$$现期量=现期量 *(1+r)$$

<h3><a name='leibi'>9.判断推理：类比推理（语义关系）</a></h3>




#### 考点一：近义关系、反义关系
- 近义关系：（近义关系 不等于 同义关系）
    - 开心：高兴
    - 言不由衷：口是心非
- 反义关系：
    - 勇敢：懦弱
    - 言而有信：言而无信
    
- 如果一级关系（近义关系/反义关系）选不出唯一答案——进行**二级辨析（感情色彩）【褒义词、贬义词、中性词】**

例题：  门客罗雀：无人问津    （）。
- A 大相径庭：不相上下
- B 众说纷纭：异口同声
- C 指鹿为马：张冠李戴
- D 发人深省：执迷不悟

答案解析：  
ABD都是反义词，题干是近义词，选C

- 门可罗雀： 门前可张网捕麻雀。形容门庭冷落
- 无人问津： 比喻没有人再来尝试做某件事或过问某件事
- **大相径庭： 指两者相差很远。也指大不一样或自相矛盾**
- 众说纷纭： 指人多嘴杂，议论纷纷
- 异口同声： 大家说的完全一致
- 指鹿为马： 比喻颠倒是非
- 张冠李戴： 比喻名实不符，弄错了对象
- 发人深省： 启发人们作深刻的思考，引起醒悟
- 执迷不悟： 固执于错误而不觉悟

例题： 雾里看花：走马观花：盲人摸象 （）
- A 小心翼翼：谨小慎微：战战兢兢
- B 兵不厌诈：智勇双全：斗智斗勇
- C 三心二意：朝三暮四：七嘴八舌
- D 虚张声势：徒有虚名：名不副实

答案解析：选D
- 题干都是近义词
- A 小心：小心：因害怕 小心 都是近义词
- B 兵不厌诈和智勇双全不是近义词；智勇双全是形容词，斗智斗勇是行为，不是近义词，排除
- C 三心二意和朝三暮四是近义词；七嘴八舌不是。排除
- D 虚张声势是假装，与实际不一样，和后面两个都是近义词。

- 二级比较：A选项和D选项 与 题干的感情色彩【贬义】
    - A和D相比较，D选项的贬义色彩更高

- 雾里看花：比喻看事情不真切
- 走马观花：骑在奔跑的马上看花。现在多指大略地观察一下，看的不深入
- 盲人摸象：这现在多比喻看问题不全面，以偏概全
- 小心翼翼：现在形容举动十分谨慎，一点不敢疏忽
- 谨小慎微：形容非常谨慎，现在多指在小事情上特别伤心
- 战战兢兢：形容极端害怕而小心谨慎的样子
- 兵不厌诈：用兵作战可以无限制地用计谋迷惑地方
- 智勇双全：既有智慧又很勇敢
- 斗智斗勇：用智谋来争胜负，凭实力分输赢
- 七嘴八舌：形容人多嘴杂
- **虚张声势：指假装除强大的声望和实力**
- 徒有虚名：突然有个虚名声而已。形容只有虚名没有真才实学的人。也影响名实不符的其它事务
- 名不副实：徒有虚名，指名声和实际不一致。

例题： 成败：呼吸  （）
- A 拉扯：拖拽
- B 好歹：始终
- C 匆忙：潇洒
- D 推荐：录用

答案解析：  选B
- 如果题干的词既不是近义词也不是反义词
- 则考虑将词 拆开来看。

- “成”和“败”；“呼”和“吸” 是互相为 反义词


- A选项 近义词
- B选项 反义词

题目：素交：故人 （）
- A 豁达：宽广
- B 遗赠：继承
- C 自由：权利
- D 颓废：消沉

答案解析： 
- 题干 素交 指真诚纯洁的友情；旧交。 和 故交 是近义关系

- A选项 豁达形容性格开朗，气量大；宽广 一般指面积或范围广大，也可以指心胸开阔，见识广博。不是近义词
- B选项  遗赠指立夏遗嘱将财产赠给别人；继承是继承财产；不是近义词
- C选项 自由是一种权利，两者是种属关系，排除
- D选项 颓废指精神萎靡，意志消沉，与消沉是近义词，选D

#### 考点二：比喻象征义
- 比喻义、象征义：
    - 把一种事务比喻成另外的食物；
    - 或者词语本身的含义同时是另外一种事务的象征

    - 例如：
        - 月亮：玉盘；
        - 松鹤：长寿
- 需要多积累

例题：蜡烛：教师 （）
- A 明月：思念
- B 桃李：学生
- C 长寿：松柏
- D 高尚：巍峨

答案解析：
- 题干：常将 教师 比喻成 蜡烛 【比喻关系】

- A选项：一般是看到明月，引发 思念。不是将 思念 比喻成 明月。排除
- B选项：一般将 学生 比喻成 桃李
- C选项：一般将 松柏 象征 长寿，但是词语的顺序反了，松柏应该和 蜡烛 对应词序，排除
- D选项：巍峨 和高尚 无明显比喻关系，排除
