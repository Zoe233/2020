# 1.事业单位是什么？

先简单了解事业单位是什么？招考的岗位有哪些？历年重要的时间节点。

## 【事业单位】  
事业单位，是指由政府利用国有资产设立的，从事教育、科技、文化、卫生等活动的社会服务组织。事业单位**接受政府领导**，是表现形式为组织或机构的法人实体。      
事业单位一般是国家设置的带有一定的公益性质的机构，但**不属于政府机构**，与公务员是不同的。        
根据国家事业单位分类改革精神，事业单位不再分为全额拨款事业单位、差额拨款事业单位，而分为公益一类事业单位、公益二类事业单位，目前还新兴了利用国有资产举办的事业单位和社会资本举办事业单位，是国家不拨款的事业单位。    


【解读】  
1. 事业单位是指由政府利用国有资产设立的，从事教育、科技、文化、卫生等活动的社会服务组织。
    - 事业单位接受政府领导，但不属于政府机构。
    - 编制和公务员不同，公务员上岸之后是行政编，事业单位叫事业编。   
2. 考公务员的时候会发现，并不是所有岗位都是行政编，也有一部分岗位是事业编 ，即**参公**。    
    - 参公和公务员一起考，编制是事业编。   
    - 参公和事业单位编制相同，但管理不同，享受公务员待遇。    
    - 公务员和事业单位的管理制度不同，分别有自己的一套管理制度。     
    - 具体哪个比较好，没有特别大的对比性，公务员制度相对会更加完善，各种福利、晋升制度会更全，事业单位可能没那么到位，但是事业单位并不代表待遇一定比公务员差，所以不能等同，因为事业单位的岗位比较多，不同的岗位情况也不同。       
3. 官方数据显示，我国接近14亿人口公务员大概是730万编制。
    - 事业单位大概是公务员的6倍，总人数接近4500万，其中包括医疗、教师岗位等特殊岗位，去掉这些特殊岗位，大概是2500万左右的岗位。
    - 所以事业单位的编制比公务员要多。
    - 所以备考的时候，不要只盯着公务员考试，公务员考完之后也可以考虑事业单位，而且考上事业单位之后也可以报考公务员。  
4. 关于事业单位，之前大家可能经常听到**全额拨款、差额拨款、自收自支**。   
    - 改革之后变了说法，叫**公益一类事业单位、公益二类事业单位**，但只是名字不同，和之前的 全额拨款、差额拨款是差不多的。   
        - 全额拨款即全由政府主导， 比如博物馆、图书馆等等，这类机构叫公益一类。   
        - 差额拨款、自收自支的可能会有自己的一部分收入，比如卫计院校、卫生服务站等，具体哪种收入高，就要具体问题具体分析。既然有些岗位有自己的一定收入，那么，今年效益好，可能奖金就会发的相对多。 
    - 事业单位在2018年改革，改革不意味着把所有岗位都变成自收自支、企业类的，也不是事业编全由政府全额拨款，因为事业单位有4500万的基数，改革不会一次到位，而且也需要存在各类岗位。  

## 【招聘岗位】   

统一公开招聘的岗位为省属事业单位中除涉密岗位外的部分**管理类岗位、通用类专业技术岗位、工勤技能类岗位**。

具体是：
- 管理类岗位，包括事业单位中除单位领导人员岗位以外的其他管理岗位。
    - 报考这类岗位不会直接做单位领导，有些类似公务员中普通的管理岗位。
    - 管理类招考的人数一直以来是最多的。
- 通用类专业技术岗位，包括事业单位中岗位等级在八级及以下的法律类、英语类、计算机类、经济类，以及部分其他类专业技术岗位（不包括教师岗位）。 
    - 其他类专业技术岗位指各部门组织公开招聘时应聘条件中没有专业要求或笔试考试专业内容在60%以下的专业技术岗位等。
    - 经济类专业技术岗位包括会计、审计、统计和其他经济岗位。
    - 大家可能对八级不理解。曾经听过关于职称的说法，正高级、副高级。通用类专业技术岗位的职称一共分为13个等级。大的分类有高级、中级、初级。高级中分为高级中分为正高1-4级、副高5-7级，中级8-10级，初级11-13级。所以专业所以专业技术类技术类招聘的招聘的岗位是中级、初级岗位。岗位是中级、初级岗位。 
- 工勤技能类岗位，包括事业单位中岗位等级在二级（技师）及以下的技术工岗位和普通工岗位。
    - 所有岗位招聘中占比最小，大概占10%。
    - 属于事业单位中的工人，分为技术工岗位和普通工岗位（不分级别）。
    - 技术工岗位 ，有职称分为5级：高级技师（一级）、技师（二级）、高级工（三级）、中级工、初级工。
    - 事业单位招考从二级（技术）及以下开始。

# 2. 事业单位考什么？

考试科目及范围？测试内容？考试题型和测试时限？

## 【考试科目及范围】
《综合知识和能力素质》

看到大纲后会发现，这几类岗位考试《综合知识和能力素质》，是一张卷，相当于行测部分和申论部分放在一起考查。
- 在事业单位中不叫行测和申论。 
- 行测全称是行政职业能力测验，是行政编
- 事业单位叫做职测，是事业编

申论，申而论之，事业单位中更多的是通过材料去测查方向性、原则性的内容，考查做事能力和解决问题的能力，而且考查得比较杂，材料没有公务员申论那么规范，所以后半部分叫综合应用能力，简称综应。

## 【测试内容】  
综合知识测试内容
- 主要测试应试人员对政治、经济、法律、管理、科技、人文等综合基础知识的掌握程度。
    - 政治（大概4-5题），可以关注社会热点、时政信息，听时政课
    - 马哲，不会特别深，相对公基会简单些。
    - 常识会考法律（3道题左右，谚语、社会现象、道理）、管理、时政。
    - 江苏是否会考公基：2015年前考公基，15年后以常识的形式测查。
    - 公基会考法律、常识也会考法律。
        - 公基是公共基础知识，考查内容更专业
        - 常识是日常生活中基本能知道的知识，也会考查的法律，就不会像公基考法律那么专业，江苏事业单位考试法律一般会有三道题：一般考法律中常用的言语，不会考具体的法律条文；通过一个现象反映出法律的哪些道理；或者法律会用最近的社会现象、最基础的常识内容去测查，反映出法律中哪些比较常用的说法。
        - 考查新法，如《慈善法》。做慈善需要申请报备，防止打着慈善的幌子揽钱
        
        

基本能力测试内容
- 主要测试应试人员阅读理解能力、判断推理能力、处理数量关系能力、综合分析能力、解决问题能力、文字表达能力，以及履行岗位职责的必备能力等素质。
    

## 【考试题型和测试时间】  
题型：单项选择题、多项选择题、简答题、论述题、综合分析题、案例分析题、事务题、材料处理题、写作题等。

测试时间：两门均150分钟，满分100分。

后面3题是综应部分，题量少，分值高，50分。

- 常识，占1/4
    - 很光，很杂，不能短时间快速提高
    - 国考的常识课、常识，时政专项课
    - 5000题中的常识的内容
    - 公基用处不大，太过专业，常识没那么专业
    
- 言语
    - 逻辑填空（实词、成语、混搭）
    - 片段阅读（中心理解，细节判断）
    - 语句表达（接语选择、语句填空）
- 判断
    - 占比最多，和江苏公务员考试类似
    - 每年约5个送分题
    - 事件排序（送分） 、类比推理、图形推理、逻辑判断、定义判断
    
- 数资
    - 多级数列、机械划分、分数数列、因式分解数列
    - 简单计算、增长率、增长量、比重、基期与现期
- 综应
    - 3道题，共50分
    - 单一题、综合题、公文题、大作文
        - 单一题（人物品质、对策做法、工作思路、不足、新现象、建议、问题）
        - 综合题（观点现象分析、对比分析）
        - 公文题（演讲稿、公开信）
        - 大作文（人物品质、自信修身、公共健康、社会治理、政府夫妇、社会公共服务）
